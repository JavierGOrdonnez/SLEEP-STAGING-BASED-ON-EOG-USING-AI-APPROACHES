# Double CNN - With Adagrad and Class weights
Let's start with the double CNN (both Fs/2 and 4Fs) and check that everything works properly.

```python
import matplotlib.pyplot as plt
import numpy as np

import sys
sys.path.append('..')
from myfunctions.myfunctions_cnn_v2 import EOG_1dcnn_cv
from myfunctions.myfunctions_cnn_v3_doubleCNN import EOG_double1dcnn_cv, EOG_fullDSN_cv
from myfunctions.myfunctions_dataloader_v3 import mydataloader_cv
```

```python
d = mydataloader_cv()
Fs = 50
root = 'D:/GitHub/TFM/'
d.load_data(path=root+'example_processed_data', K=5, max_n_files=20)
```

## 1e-2

```python
epochs = 100
batch_size = 500
lr = 1e-2
class_weights=[1, 10, 0.8, 2, 3]
savepath = root + '02_1DCNN_EOG/trained_models/doubleCNN/doubleCNN_vanilla_adagrad_1e-2_classweights/'
cnn = EOG_double1dcnn_cv(K=5, mode='vanilla',lr=lr, optimizer='adagrad',class_weights=class_weights,
                         savepath=savepath, use_GPU=True)
```

```python
for k in range(d.K):
    print('##########################################################################################')
    print('          Starting fold %d' %k)
    print('##########################################################################################')
    cnn.trainloop(d,k,epochs=epochs,restart=True,batch_size=batch_size)
results = cnn.plot_training_information()
print('DONE!')
```
