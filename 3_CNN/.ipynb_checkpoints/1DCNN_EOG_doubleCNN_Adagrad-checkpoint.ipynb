```python
import matplotlib.pyplot as plt
import numpy as np

import sys
sys.path.append('..')
from myfunctions_CNN import EOG_1dcnn_cv
from myfunctions_doubleCNN import EOG_double1dcnn_cv
from myfunctions_dataloader import mydataloader_cv
```

```python
d = mydataloader_cv()
Fs = 50
root = 'C:/Users/Javi/Documents/GitHub/TFM/'
d.load_data(path=root+'example_processed_data', K=5, max_n_files=20)
```

```python
epochs = 30
batch_size = 500
lr = 1e-2
savepath = root + '02_1DCNN_EOG/trained_models/doubleCNN/doubleCNN_vanilla_adagrad_1e-2/'

cnn = EOG_double1dcnn_cv(K=5, mode='vanilla',lr=lr, optimizer='adagrad',
                         savepath=savepath, use_GPU=True)
```

```python
for k in range(d.K):
    print('##########################################################################################')
    print('          Starting fold %d' %k)
    print('##########################################################################################')
    cnn.trainloop(d,k,epochs=epochs,restart=True,batch_size=batch_size)
print('DONE!')
```

```python
results = cnn.plot_training_information();
```
